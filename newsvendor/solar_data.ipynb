```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

```python
#year, month, day, hour3, house_id4, hourly_kwh5, hourly_solar_kwh6, hourly_gal,latitude8, longitude9, summary10, temperature11, humidity, visibility, apparent_temperature, pressure, wind_speed, cloud_cover,wind_bearing, precip_intensity, precip_probability
preprocessed_data = genfromtxt('../data/preprocessed_data.csv', delimiter=',', dtype = str)
```

```python
preprocessed_data.shape
```

```python
pre_data_s = preprocessed_data[0:10000]
pre_data_s[1:10,5]
```

```python
pre_data_s[1:10,6]
```

```python
x = np.arange(0,20)
plt.legend(["usage","solar"])
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(5)
ax1 = fig.add_subplot(211)
ax1.scatter(x,pre_data_s[0:20,5])
ax1.scatter(x,pre_data_s[0:20,6])
ax1.yaxis.set_ticks(np.arange(0, 5, 2))
ax1.set_xlabel('FIPS index')
ax1.set_ylabel('call in minutes')
ax1.legend(["min travel time", "average of 5 closest"])
plt.show()
```

```python
#2000 - 2100
x = np.arange(0,100)
lower = 2200
upper = 2300
plt.scatter(x,pre_data_s[lower:upper,5])
plt.scatter(x,pre_data_s[lower:upper,6])
plt.legend(["elec usage", "solar"])
plt.show()
```

```python
y1 = pre_data_s[lower:upper,5].astype(float)
y2 = pre_data_s[lower:upper,6].astype(float)
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.legend(["elec usage", "solar"])
plt.show()
```
